```python
# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 5 — Convert to YOLO Format                        ║
# ╚══════════════════════════════════════════════════════════╝
#
#  Annotation format (CONFIRMED):
#    parts[0] = filename  (skip)
#    parts[1] = class name
#    parts[2..5] = x1 y1 x2 y2  (absolute pixels, corner format)
#
import cv2, shutil, random, glob
from collections import Counter
from pathlib import Path

CLASS_NAMES = [
    "Folding_Knife",     # 0
    "Straight_Knife",    # 1  ← minority
    "Utility_Knife",     # 2
    "Multi-tool_Knife",  # 3
    "Scissor",           # 4
]
CLASS_MAP = {name: idx for idx, name in enumerate(CLASS_NAMES)}

def parse_annotation(ann_path, img_w, img_h):
    """
    Parse one OPIXray annotation file.
    Format: filename.jpg  ClassName  x1  y1  x2  y2
    Returns list of (class_id, cx_norm, cy_norm, w_norm, h_norm)
    """
    boxes = []
    with open(ann_path) as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) < 6:
                continue
            # parts[0] = filename (skip)
            # parts[1] = class name
            cls_raw = parts[1]
            cid = CLASS_MAP.get(cls_raw)
            if cid is None:
                # Try case-insensitive match
                for name, idx in CLASS_MAP.items():
                    if name.lower() == cls_raw.lower():
                        cid = idx
                        break
            if cid is None:
                print(f"  ⚠️  Unknown class '{cls_raw}' in {ann_path}")
                continue

            x1, y1, x2, y2 = float(parts[2]), float(parts[3]), \
                               float(parts[4]), float(parts[5])

            if x2 <= x1 or y2 <= y1:
                continue  # degenerate box

            cx = ((x1 + x2) / 2) / img_w
            cy = ((y1 + y2) / 2) / img_h
            w  = (x2 - x1) / img_w
            h  = (y2 - y1) / img_h

            # Clamp
            cx = min(max(cx, 0.0), 1.0)
            cy = min(max(cy, 0.0), 1.0)
            w  = min(max(w,  0.0), 1.0)
            h  = min(max(h,  0.0), 1.0)

            boxes.append((cid, cx, cy, w, h))
    return boxes


def gather_pairs(img_dir, ann_dir, label):
    pairs, missing, bad = [], 0, 0
    for img_path in sorted(Path(img_dir).glob("*.jpg")):
        stem     = img_path.stem
        ann_path = Path(ann_dir) / f"{stem}.txt"
        if not ann_path.exists():
            missing += 1
            continue
        img = cv2.imread(str(img_path))
        if img is None:
            bad += 1
            continue
        ih, iw = img.shape[:2]
        boxes = parse_annotation(str(ann_path), iw, ih)
        if boxes:
            pairs.append((str(img_path), boxes))
    print(f"[{label}] pairs={len(pairs)}  missing_ann={missing}  bad_img={bad}")
    return pairs


FLAG2 = f"{WORK_DIR}/.converted"
if Path(FLAG2).exists():
    print("✅  Already converted — skipping")
    # Still load CLASS_NAMES for later cells
else:
    print("Parsing annotations…")

    train_all = gather_pairs(
        f"{RAW_DIR}/train/train_image",
        f"{RAW_DIR}/train/train_annotation",
        "TRAIN+VAL"
    )
    test_pairs = gather_pairs(
        f"{RAW_DIR}/test/test_image",
        f"{RAW_DIR}/test/test_annotation",
        "TEST"
    )

    # Carve 15% val out of train
    random.seed(42)
    random.shuffle(train_all)
    n_val       = max(1, int(0.15 * len(train_all)))
    val_pairs   = train_all[:n_val]
    train_pairs = train_all[n_val:]

    splits = {"train": train_pairs, "val": val_pairs, "test": test_pairs}

    print("\n── Split sizes ──────────────────────────────────────")
    for sp, pairs in splits.items():
        print(f"  {sp:5s}: {len(pairs)} images")

    print("\n── Class distribution (train) ───────────────────────")
    cnt = Counter()
    for _, boxes in train_pairs:
        for (cid, *_) in boxes:
            cnt[cid] += 1
    for i, name in enumerate(CLASS_NAMES):
        bar = "█" * (cnt[i] // 40)
        print(f"  [{i}] {name:20s}: {cnt[i]:5d}  {bar}")

    print("\nWriting YOLO dataset…")
    for sp, pairs in splits.items():
        img_out = Path(f"{DATASET_DIR}/images/{sp}")
        lbl_out = Path(f"{DATASET_DIR}/labels/{sp}")
        img_out.mkdir(parents=True, exist_ok=True)
        lbl_out.mkdir(parents=True, exist_ok=True)

        for img_path, boxes in pairs:
            fname = os.path.basename(img_path)
            stem  = os.path.splitext(fname)[0]
            shutil.copy2(img_path, img_out / fname)
            with open(lbl_out / f"{stem}.txt", "w") as f:
                for (cid, cx, cy, w, h) in boxes:
                    f.write(f"{cid} {cx:.6f} {cy:.6f} {w:.6f} {h:.6f}\n")

        print(f"  ✅  {sp:5s} written")

    Path(FLAG2).touch()
    print("\n✅  Conversion complete")
```

```python
# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 6 — Write dataset.yaml                            ║
# ╚══════════════════════════════════════════════════════════╝
import yaml

YAML_PATH = f"{WORK_DIR}/opixray.yaml"
cfg = {
    "path":  DATASET_DIR,
    "train": "images/train",
    "val":   "images/val",
    "test":  "images/test",
    "nc":    len(CLASS_NAMES),
    "names": CLASS_NAMES,
}
with open(YAML_PATH, "w") as f:
    yaml.dump(cfg, f, default_flow_style=False)

print("dataset.yaml:")
print(open(YAML_PATH).read())
```

```python

# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 7 — Compute Class Weights                         ║
# ╚══════════════════════════════════════════════════════════╝
import numpy as np
from collections import Counter
from pathlib import Path

cnt = Counter()
for lf in Path(f"{DATASET_DIR}/labels/train").glob("*.txt"):
    with open(lf) as f:
        for line in f:
            cid = int(line.split()[0])
            cnt[cid] += 1

total = sum(cnt.values())
freqs   = np.array([cnt.get(i, 1) for i in range(len(CLASS_NAMES))], dtype=float)
weights = (total / (len(CLASS_NAMES) * freqs))
weights = weights / np.median(weights)

print("Class weights (upweights minority Straight_Knife):")
for i, (name, w) in enumerate(zip(CLASS_NAMES, weights)):
    print(f"  [{i}] {name:20s}  cnt={cnt[i]:5d}  weight={w:.3f}")

MAX_CLS_WEIGHT = float(weights.max())
print(f"\ncls_pw = {MAX_CLS_WEIGHT:.3f}  (passed to YOLOv8 trainer)")

```

```python
# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 8 (FIXED) — TRAIN                                 ║
# ╚══════════════════════════════════════════════════════════╝
from ultralytics import YOLO
import os, shutil

RUNS_DIR     = "/content/opixray/runs"
YAML_PATH    = "/content/opixray/opixray.yaml"
RUN_DIR      = f"{RUNS_DIR}/opixray_v8m"
LAST_WEIGHTS = f"{RUN_DIR}/weights/last.pt"
BEST_WEIGHTS = f"{RUN_DIR}/weights/best.pt"
DRIVE_BEST   = "/content/drive/MyDrive/Colab Notebooks/opixray_best.pt"

if os.path.exists(LAST_WEIGHTS):
    print(f"⚡  Resuming from checkpoint: {LAST_WEIGHTS}")
    model   = YOLO(LAST_WEIGHTS)
    results = model.train(resume=True)

else:
    print("🚀  Fresh training — yolov8m.pt")
    model = YOLO("yolov8m.pt")
    results = model.train(
        data         = YAML_PATH,
        epochs       = 100,
        imgsz        = 960,
        batch        = -1,        # auto-batch to fill GPU VRAM
        optimizer    = "AdamW",
        lr0          = 1e-3,
        lrf          = 0.01,
        weight_decay = 0.0005,
        warmup_epochs= 3,
        # ── Loss ───────────────────────────────────────────
        box          = 7.5,
        cls          = 0.5,
        dfl          = 1.5,
        # ── Augmentation (handles class imbalance) ─────────
        hsv_h        = 0.015,
        hsv_s        = 0.7,
        hsv_v        = 0.4,
        degrees      = 10.0,
        translate    = 0.1,
        scale        = 0.5,
        fliplr       = 0.5,
        mosaic       = 1.0,       # key for occlusion
        mixup        = 0.15,
        copy_paste   = 0.3,       # synthetic occlusion
        # ── Runtime ────────────────────────────────────────
        device       = 0,
        workers      = 4,
        patience     = 25,
        amp          = True,
        save         = True,
        save_period  = 5,
        project      = RUNS_DIR,
        name         = "opixray_v8m",
        exist_ok     = True,
        plots        = True,
        val          = True,
        verbose      = True,
    )

shutil.copy2(BEST_WEIGHTS, DRIVE_BEST)
print(f"\n✅  Best weights saved to Drive: {DRIVE_BEST}")
```

```python

# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 9 — Evaluate on Held-Out Test Set                 ║
# ╚══════════════════════════════════════════════════════════╝
import json
from ultralytics import YOLO

model   = YOLO(BEST_WEIGHTS)
metrics = model.val(
    data     = YAML_PATH,
    split    = "test",
    imgsz    = 960,
    batch    = 8,
    device   = 0,
    plots    = True,
    project  = RUNS_DIR,
    name     = "opixray_eval",
    exist_ok = True,
)

print("\n════ TEST SET RESULTS ════════════════════════════════")
print(f"  mAP50     : {metrics.box.map50:.4f}")
print(f"  mAP50-95  : {metrics.box.map:.4f}")
print(f"  Precision : {metrics.box.mp:.4f}")
print(f"  Recall    : {metrics.box.mr:.4f}")

print("\n  Per-class:")
results_dict = {}
for i, name in enumerate(CLASS_NAMES):
    p   = float(metrics.box.p[i])
    r   = float(metrics.box.r[i])
    ap  = float(metrics.box.ap50[i])
    fn  = 1 - r
    ok  = "✅" if r >= 0.80 and fn <= 0.10 else "⚠️  BELOW TARGET"
    print(f"  [{i}] {name:20s}  P={p:.3f}  R={r:.3f}  AP50={ap:.3f}  FN={fn:.3f}  {ok}")
    results_dict[name] = {"precision": p, "recall": r, "AP50": ap, "FN_rate": fn}

# Save JSON to Drive
out = {
    "mAP50": float(metrics.box.map50),
    "mAP50_95": float(metrics.box.map),
    "per_class": results_dict,
}
json_path = "/content/drive/MyDrive/Colab Notebooks/opixray_results.json"
with open(json_path, "w") as f:
    json.dump(out, f, indent=2)
print(f"\n✅  Results saved to Drive: {json_path}")
```

```python

# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 10 — Visual Sample Predictions                    ║
# ╚══════════════════════════════════════════════════════════╝
import cv2, random, glob
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from ultralytics import YOLO

model       = YOLO(BEST_WEIGHTS)
test_images = glob.glob(f"{DATASET_DIR}/images/test/*.jpg")
sample      = random.sample(test_images, min(6, len(test_images)))

COLOURS = ["#e74c3c","#9b59b6","#2980b9","#27ae60","#f39c12"]

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
for ax, img_path in zip(axes.flatten(), sample):
    img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
    preds = model.predict(img_path, imgsz=960, conf=0.25, verbose=False)
    ax.imshow(img)
    for box in preds[0].boxes:
        cid  = int(box.cls[0])
        conf = float(box.conf[0])
        x1, y1, x2, y2 = box.xyxy[0].tolist()
        ax.add_patch(patches.Rectangle(
            (x1, y1), x2-x1, y2-y1,
            linewidth=2, edgecolor=COLOURS[cid], facecolor="none"
        ))
        ax.text(x1, y1-5, f"{CLASS_NAMES[cid]} {conf:.2f}",
                color="white", fontsize=7,
                bbox=dict(facecolor=COLOURS[cid], alpha=0.8, pad=1))
    ax.axis("off")
    ax.set_title(os.path.basename(img_path), fontsize=8)

plt.tight_layout()
out_png = "/content/drive/MyDrive/Colab Notebooks/opixray_predictions.png"
plt.savefig(out_png, dpi=150, bbox_inches="tight")
plt.show()
print(f"✅  Saved to Drive: {out_png}")
```

```python
# integration_api.py  — save this to Drive
from ultralytics import YOLO
import base64, json, tempfile, os

CLASS_NAMES = ["Folding_Knife","Straight_Knife","Utility_Knife","Multi-tool_Knife","Scissor"]
model = YOLO("/content/drive/MyDrive/Colab Notebooks/opixray_best.pt")

def detect(image_input, conf=0.25):
    """
    Args: image_input = file path OR base64 string
    Returns: JSON list of detections
    """
    if not os.path.exists(str(image_input)):
        img_bytes = base64.b64decode(image_input)
        with tempfile.NamedTemporaryFile(suffix=".jpg", delete=False) as f:
            f.write(img_bytes)
            image_input = f.name

    results = model.predict(image_input, imgsz=960, conf=conf, verbose=False)
    detections = []
    for box in results[0].boxes:
        detections.append({
            "class": CLASS_NAMES[int(box.cls[0])],
            "confidence": round(float(box.conf[0]), 3),
            "bbox_normalised": {
                "cx": round(float(box.xywhn[0][0]), 4),
                "cy": round(float(box.xywhn[0][1]), 4),
                "w":  round(float(box.xywhn[0][2]), 4),
                "h":  round(float(box.xywhn[0][3]), 4),
            }
        })
    return json.dumps({"detections": detections, "count": len(detections)}, indent=2)

# Test it
import glob, random
test_img = random.choice(glob.glob("/content/opixray/dataset/images/test/*.jpg"))
print(detect(test_img))
```
